```python
#updated code
import pandas as pd
```

```python
file_path="/content/drive/MyDrive/question_1.xlsx"
df= pd.read_excel(file_path,engine="openpyxl")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df["total"]=df[["math score","reading score","writing score"]].sum(axis=1)
df["maximum"]=df[["math score","reading score","writing score"]].max(axis=1)
df["minimum"]=df[["math score","reading score","writing score"]].min(axis=1)
df["average"]=df[["math score","reading score","writing score"]].mean(axis=1)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
df["total"].plot(kind='hist',bins=10,color="skyblue",edgecolor="magenta",title="histogram of total")
plt.xlabel("total marks")
plt.ylabel("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
df["math score"].plot(kind='hist',bins=10,color="cyan",edgecolor="red",title="histogram of math score")
plt.xlabel("marks")
plt.ylabel("frequency")
plt.show()
```
